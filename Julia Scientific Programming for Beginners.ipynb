```julia
#csvtt
loc = ".\\Julia_Scientific_Programming_for_Beginners\\"
filename = "temp_2019_tokyo.csv"
file = open(loc*filename)

#csv
month = zeros(Int64, 0)
temperature = zeros(Float64, 0)
precipitation = zeros(Float64, 0)

#1String
#eachline
i = 0
for line in eachline(file)
    i += 1
    #1header
    i == 1 && continue
    #(csv)
    s = split(line, ",")
    #s3String
    m = parse(Int64, s[1])
    push!(month, m)
    t = parse(Float64, s[2])
    push!(temperature, t)
    p = parse(Float64, s[3])
    push!(precipitation, p)
end 
   
println(month)
println(temperature)
println(precipitation)
close(file)
```
